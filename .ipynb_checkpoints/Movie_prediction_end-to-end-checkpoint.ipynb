13.

Fit 3 linear regression models on the training data:
•
Model 1: Use only runtime_in_minutes
•
Model 2: Use runtime_in_minutes and kid_friendly
•
Model 3: Use runtime_in_minutes, kid_friendly and the dummy columns for the genres

```python
y_train = df[df['in_theaters_date'].dt.year < 2010]['critic_rating']
y_test=  df[df['in_theaters_date'].dt.year >= 2010]['critic_rating']

X_train_model_1 = features_train['runtime_in_minutes'].values.reshape(-1, 1)
X_train_model_2 = features_train[['runtime_in_minutes','kid_friendly']]
X_train_model_3 = features_train.drop(columns=['movie_title'])
```

```python
import statsmodels.api as sm

X_stats_train_model1 = sm.add_constant(X_train_model_1)
X_stats_train_model2 = sm.add_constant(X_train_model_2)
X_stats_train_model3 = sm.add_constant(X_train_model_3)


# Fit the OLS model
model1 = sm.OLS(y_train, X_stats_train_model1).fit()
model2 = sm.OLS(y_train, X_stats_train_model2).fit()
model3 = sm.OLS(y_train, X_stats_train_model3).fit()
```

14. Score the linear regression models on the test data by writing a function where you can input the y_test and y_pred values (y_pred = predicted values after you apply the fitted model to your X_test data), and it outputs the following metrics: R2, MAE and RMSE. Apply the function to the three models that you’ve fit so far.

```python
def evaluate_model_statsmodel(y_test, y_pred):    
    # Calculate residuals
    residuals = y_test - y_pred
    
    # Calculate R^2
    r2 = 1 - (np.sum(residuals**2) / np.sum((y_test - np.mean(y_test))**2))
    
    # Calculate MAE
    mae = np.mean(np.abs(residuals))
    
    # Calculate RMSE
    rmse = np.sqrt(np.mean(residuals**2))
    
    return r2, mae, rmse


```

```python
new_features = test[['movie_title', 'runtime_in_minutes']].copy()
kid_friendly = test['rating'].isin(['G', 'PG']).astype(int).to_frame()
dummy_genres = test['genre'].str.get_dummies(', ')
features_test = pd.concat([new_features, kid_friendly,dummy_genres], axis=1)
features_test.rename(columns={'rating': 'kid_friendly'}, inplace=True)
X_test_model_1 = features_test['runtime_in_minutes'].values.reshape(-1, 1)
X_test_model_2 = features_test[['runtime_in_minutes','kid_friendly']]
X_test_model_3 = features_test.drop(columns=['movie_title'])
```

```python
X_stats_test_model1 = sm.add_constant(X_test_model_1)
X_stats_test_model2 = sm.add_constant(X_test_model_2)
X_stats_test_model3 = sm.add_constant(X_test_model_3)

y_pred_model_1 = model1.predict(X_stats_test_model1)
y_pred_model_2 = model2.predict(X_stats_test_model2)
y_pred_model_3 = model3.predict(X_stats_test_model3)
```

```python
evaluate_model_statsmodel(y_test, y_pred_model_1)
```

```python
evaluate_model_statsmodel(y_test, y_pred_model_2)
```

```python
evaluate_model_statsmodel(y_test, y_pred_model_3)
```
